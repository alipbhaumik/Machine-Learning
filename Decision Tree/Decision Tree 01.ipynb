```python
#Alternative implementation
import seaborn as  sns
```

```python
iris=sns.load_dataset("iris")
iris
```

```python
#Here the labels are in proper string format 
#We then use LabelEncoder from scikit-learn to encode the 'species' column into numerical labels 
#and add it as a new column 'species_encoded'.
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
iris["species_encoded"]=label_encoder.fit_transform(iris["species"])
```

```python
iris.head()
```

```python
iris.shape
```

```python
#Independent features
X= iris.iloc[:,:-2]
X.head()
```

```python
#Dependent feature
y=iris.iloc[:,-1]
y.head()
```

```python
X,y
```

```python
#Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

treemodel= DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
```

```python
tree.plot_tree(treemodel,filled=True)
```

```python
#For post-pruning we will tickle with hyper-parameters such as max_depth
treemodel= DecisionTreeClassifier(max_depth=2)
treemodel.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score= accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
#Pre - Pruning
```

```python
parameter = {
    'criterion':["gini","entropy","log_Loss"],
    "splitter":["best","random"],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
}
```

```python
from sklearn.model_selection import GridSearchCV
#GridSearchCV, or Grid Search Cross-Validation, 
#is a machine learning technique that finds the best parameter values
#in a given set of parameters. 
#It's a cross-validation technique
#that involves entering a model and parameters,
#extracting the best values, and then making predictions. 
```

```python
treemodel= DecisionTreeClassifier()
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score= accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
